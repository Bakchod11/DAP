# Task 1

```python
BOOKS = []

def insert():
    name = input("Enter the book name: ")
    author = input("Enter the author: ")
    price = int(input("Enter the price: "))
    copies = int(input("Enter the copies: "))
    book = {"name": name, "author": author, "price": price, "copies": copies}
    BOOKS.append(book);

def display(book):
    print(f"Book: {book['name']}")
    print(f"Author: {book['author']}")
    print(f"Price: {book['price']}")
    print(f"Total Copies: {book['copies']}")
    print("--------------------------------------")

def display_all():
    for book in BOOKS:
        print(f"Book: {book['name']}")
        print(f"Author: {book['author']}")
        print(f"Price: {book['price']}")
        print(f"Total Copies: {book['copies']}")
        print("-----------------------------------")

def search_by_name():
    book_name = input("Enter the book name: ")
    book_found = {}
    for book in BOOKS:
        if(book["name"] == book_name):
            book_found = book
        if book_found:
            display(book_found)
        else:
            print("No book found")

def search_by_author():
    author_name = input("Enter the author name: ")
    for book in BOOKS:
        if(book["author"] == author_name):
            display(book)

def sort_by_name():
    newList = sorted(BOOKS, key= lambda d:d['name'])
    return newList

def main():
    choice = 0
    while(choice != 6):
        print("Welcome to library. What would you like to do? ")
        print("Enter your choice: ")
        print("1. Insert a book")
        print("2. Display all books")
        print("3. Search by book name")
        print("4. Search by author name")
        print("5. Sort the List")
        print("6. Exit")
        choice = int(input())

        if(choice == 1):
            insert()
        elif(choice == 2):
            display_all()
        elif(choice == 3):
            search_by_name()
        elif(choice == 4):
            search_by_author()
        elif(choice == 5):
            sort_by_name()
            display_all()
        else:
            print("Enter a valid choice")

main()
```

# Task 2

```python
EMPLOYEES = [
    {
        "id": 101,
        "name": "Jack",
        "department": "HR",
        "payroll": "weekly",
    },
    {
        "id": 102,
        "name": "Anita",
        "department": "Marketing",
        "payroll": "monthly",
        "salary": 20000,
    },
    {
        "id": 103,
        "name": "John",
        "department": "Sales",
        "payroll": "weekly",
        "salary": 3500,
    },
]

def insert():
    id = int(input("Enter the id: "))
    name = input("Enter the employee name: ")
    department = input("Enter the department: ")
    payroll = input("Enter the payroll type: ")
    salary = int(input("Enter the salary: "))
    employee = {"id": id, "name": name, "department": department, "payroll": payroll, "salary": salary}
    EMPLOYEES.append(employee);

def weekly_wage_bill():
    id = int(input("Enter the employee id: "))
    for emp in EMPLOYEES:
        if(emp["id"] == id):
            if(emp["payroll"] == "weekly"):
                print(f"The weekly wage bill is {emp['salary']}")
            else:
                weekly_wage = (emp["salary"] * 12) / 52
                # print(f"The weekly wage bill is {emp['salary']/4}")
                print(f"The weekly wage bill is {round(weekly_wage,2)}")

def monthly_wage_bill():
    id = int(input("Enter the employee id: "))
    for emp in EMPLOYEES:
        if(emp["id"] == id):
            if(emp["payroll"] == "monthly"):
                print(f"The monthly wage bill is {emp['salary']}")
            else:
                monthly_wage = (emp["salary"] * 52) / 12
                print(f"The monthly wage bill is {round(monthly_wage,2)}")

def yearly_wage_bill():
    id = int(input("Enter the employee id: "))
    for emp in EMPLOYEES:
        if(emp["id"] == id):
            if(emp["payroll"] == "weekly"):
                print(f"The yearly wage bill is {emp['salary'] * 52}")
            else:
                print(f"The yearly wage bill is {emp['salary'] * 12}")

def payroll_means():
    weekly = 0
    monthly = 0
    for emp in EMPLOYEES:
        if(emp["payroll"] == "weekly"):
            weekly += 1
        else:
            monthly+= 1
    print(f"The employees paid monthly are {monthly} and weekly are {weekly}")

def main():
    choice = 0
    while(choice != 6):
        print("Enter your choice: ")
        print("1. Insert employee record")
        print("2. Employee weekly wage bill")
        print("3. Employee monthly wage bill")
        print("4. Employee yearly wage bill")
        print("5. Number of employees being paid by each means.")
        print("6. Exit")
        choice = int(input())

        if(choice == 1):
            insert()
        elif(choice == 2):
            weekly_wage_bill()
        elif(choice == 3):
            monthly_wage_bill()
        elif(choice == 4):
            yearly_wage_bill()
        elif(choice == 5):
            payroll_means()
        else:
            print("Enter a valid choice")

if __name__ == "__main__":
    main()
```

```python
# Employee records
employees = [
    {"name": "John Doe", "department": "IT", "id": 123, "salary_type": "weekly", "amount": 500},
    {"name": "Jane Doe", "department": "HR", "id": 124, "salary_type": "monthly", "amount": 6000},
    {"name": "Jim Smith", "department": "IT", "id": 125, "salary_type": "weekly", "amount": 700},
    {"name": "Sarah Johnson", "department": "Marketing", "id": 126, "salary_type": "monthly", "amount": 7000}
]

# Initialize variables to store weekly and monthly wages
weekly_wage_bill = 0
monthly_wage_bill = 0

# Initialize variables to store employee count by salary type
weekly_salary_count = 0
monthly_salary_count = 0

# Loop through the employees list to process each record
for employee in employees:
    if employee["salary_type"] == "weekly":
        weekly_wage_bill += employee["amount"]
        weekly_salary_count += 1
    else:
        monthly_wage_bill += employee["amount"]
        monthly_salary_count += 1

# Calculate the total yearly wage bill
yearly_wage_bill = weekly_wage_bill * 52 + monthly_wage_bill * 12

# Output the results
print("Weekly wage bill:", weekly_wage_bill)
print("Monthly wage bill (excluding weekly wages):", monthly_wage_bill)
print("Total yearly wage bill:", yearly_wage_bill)
print("Number of employees with weekly salary:", weekly_salary_count)
print("Number of employees with monthly salary:", monthly_salary_count)
```
